```python
import os
import cv2

vid_path = '../db/original/'
pic_path = '../db/pic/'
pic_name = 'Brien_R/'
vid_name = 'Brien_R.mp4'

try: 
    if not(os.path.isdir(pic_path + pic_name)):
        os.makedirs(os.path.join(pic_path + pic_name))
except OSError as e:
    if e.errno != errno.EEXIST:
        print("Failed to create directory")
        raise

        
vidcap = cv2.VideoCapture(vid_path + vid_name)
        
cnt = 1
success = True
        
while success :
    success, img = vidcap.read()
    cv2.imwrite(pic_path + pic_name + '%d.jpg' % cnt, img)
    print("saved image %d.jpg" % cnt)
    
    if cv2.waitKey(10) == 27:
        break
        
    cnt += 1
```

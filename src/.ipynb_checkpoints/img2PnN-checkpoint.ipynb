```python
import re
from PIL import Image
from PIL import ImageDraw


positive_path = '../positive/'
negative_path = '../negative/'
cnt = 1

with open('../pupil.txt', 'r') as f :
    for line in f :
        info = line.split(' ')
        
        img_path = info[0]
        print(img_path)
        
        x = int(info[2])
        y = int(info[3])
        
        w = int(info[4])
        h = int(info[5])
        
        ROI = (x, y, x+w, y+h)
        
        img = Image.open(img_path)
        print(ROI)
        print(img.size)
        
        cropped_img = img.crop(ROI)
        
        masked_img = img
        draw = ImageDraw.Draw(masked_img)
        draw.rectangle(((x,y),(x+w,y+h)),fill=0)
        
        
        # cropped_img.show()
        masked_img.save(negative_path + '0831N_' + str(cnt) + '.jpg')
        cropped_img.save(positive_path + '0831P_' + str(cnt) + '.jpg')
        
        cnt += 1
        
        
        
        

        
```
